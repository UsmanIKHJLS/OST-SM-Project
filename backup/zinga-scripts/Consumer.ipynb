```python
#!pip install kafka
from confluent_kafka import Consumer, KafkaException
import json

# Kafka Consumer Configuration
consumer_conf = {
    'bootstrap.servers': 'kafka:9092',  # Kafka address
    'group.id': 'consumer-group-1',     # Consumer group name
    'auto.offset.reset': 'earliest'    # Start from the beginning of the topic if there are no commits
}

# Create the consumer
consumer = Consumer(consumer_conf)
topic = 'kf-topic'
consumer.subscribe([topic])

print(f"Kafka consumer connected and listening to the topico '{topic}'...")

try:
    while True:
        msg = consumer.poll(1.0)  # Wait for message for up to 1 second
        if msg is None:
            continue
        if msg.error():
            if msg.error().code() == KafkaException._PARTITION_EOF:
                # End of partition
                continue
            else:
                raise KafkaException(msg.error())
        # Decode a message
        try:
            payload = json.loads(msg.value().decode('utf-8'))
            print(f"Received mesaage : {payload}")
        except json.JSONDecodeError as e:
            print(f"Error to Decode the message : {msg.value()}. Details: {e}")
except KeyboardInterrupt:
    print("\nStoping the consumer...")
finally:
    # Close the consumer and leave 
    consumer.close()
```
